````python
prompts = [
    """
        You are a course researcher for Krux Learn.

        Your job is to research a course topic and extract only the most useful, practical, and actionable information for working professionals.
        
        Course title:
        Set up Claude as your marketing assistant
        
        Your task:
        Research this topic deeply using:
        1. Official company sources first (product pages, help docs, guides, release notes, best-practice documentation)
        2. Credible practitioner usage second (case studies, blog posts, workflows, videos, examples, public discussions, community posts)
        
        Goal:
        Find out:
        - what the product/tool actually is
        - which features are relevant to this course
        - how marketers are actually using it in real life
        - what repeatable workflows are most valuable
        - what setup steps, templates, and inputs are needed
        - what common mistakes people make
        - what advice would make this genuinely useful for someone trying to apply it today
        
        Research principles:
        - Prioritize official sources for facts and feature accuracy
        - Use practitioner sources to understand real-world usage
        - Do not include fluff, generic AI advice, or vague benefits
        - Focus only on information relevant to the course topic
        - Prefer workflows, examples, templates, and step-by-step actions over theory
        - Highlight anything that is especially useful for marketers
        - If information is uncertain or conflicting, mention that clearly
        - Exclude highly technical details unless they directly improve usage for marketers
        
        Output format:
        Structure the output under clear sub-headings and bullet points.
        
        Use this format:
        
        1. What this course is about
        - Explain in simple terms what “Set up Claude as your marketing assistant” means
        - Clarify the core user outcome
        
        2. Relevant Claude features for this course
        - List only the features that matter for this use case
        - Explain each in 1–2 lines
        - Mention why it matters for marketers
        
        3. How marketers are using Claude in practice
        - Summarize real-world use cases
        - Group similar use cases together
        - Focus on repeatable, job-relevant workflows
        
        4. Best workflows to teach in this course
        - List the most useful workflows this course should cover
        - For each workflow, mention:
          - what the user is trying to do
          - what inputs they need
          - what Claude helps with
          - what output they get
        
        5. Recommended setup for a marketer
        - What documents, inputs, or context should a marketer prepare?
        - What should be uploaded, written, or configured?
        - What setup advice is most practical?
        
        6. Prompts, instructions, or templates worth including
        - Extract useful examples or patterns
        - Summarize them as reusable guidance
        - Do not dump random prompts without context
        
        7. Common mistakes and limitations
        - What usually goes wrong?
        - Where does Claude give weak outputs?
        - What should marketers avoid?
        
        8. Practical takeaways for course creation
        - What are the most important lessons to teach?
        - What should be included in the course?
        - What should be excluded?
        - What would make this course immediately actionable?
        
        Final instruction:
        Write clearly, practically, and with the mindset of building a short, high-value lesson for busy marketers.
        The output should feel like research notes that can directly be used to create a course lesson, not like a general article.
        """,
    """ 
        You are a course researcher for KRUX, a platform that teaches 
        marketers how to use AI tools in their daily work.
        
        TASK:
        Research the topic below and produce a structured course outline 
        that a content writer can turn into 100-word educational cards.
        
        TOPIC: "Set Up Claude as Your Marketing Assistant"
        
        RESEARCH REQUIREMENTS:
        1. Go to the official source (in this case, Anthropic's documentation 
           and Claude's product pages) and find exactly how the feature works.
        2. Search for real practitioners — marketers, growth leads, content 
           teams — who have shared how they use this in their actual work. 
           Look on X, LinkedIn, newsletters (especially MKT1, Future of 
           Marketing), and marketing blogs.
        3. For each sub-topic, answer:
           - What specific problem does this solve for a marketer?
           - What are the exact steps to set it up?
           - What does the before vs after look like in their workflow?
           - Are there any gotchas, limits, or common mistakes?
        
        OUTPUT FORMAT:
        For each sub-topic, provide:
        
        **Sub-topic title**
        - Problem it solves: [one line]
        - How it works: [3-5 concrete steps from official docs]
        - Real-world example: [how an actual marketer/team uses this, 
          with source if available]
        - Before vs After: [what changes in their Monday morning]
        - Common mistakes: [what NOT to do]
        - Sources: [URLs]
        
        TARGET AUDIENCE:
        Marketing managers and directors who use ChatGPT casually but 
        have never set up persistent context, skills, or projects in 
        any AI tool. Assume zero technical knowledge. They know 
        marketing deeply but treat AI as a text box they type into.
    """
]
````

````python
def openai_learn():

    output = []

    for i in prompts:
        response = client_openai.responses.create(
            model = "gpt-5.4",
            tools = [
                {
                    "type": "web_search"
                }
            ],
            include = ["web_search_call.action.sources"],
            input = i
        )

        output_text = response.output_text

        output.append(output_text)


    return print(output)
````

````python
openai_learn()
````
