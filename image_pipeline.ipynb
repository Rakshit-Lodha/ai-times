```python
from openai import OpenAI
```

```python
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
import os
```

```python
client = OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
from supabase import create_client, Client
```

```python
supabase_url = os.environ.get("SUPABASE_URL")
```

```python
supabase_api_key = os.environ.get("SUPBASE_KEY")
```

```python
supabase: Client = create_client(supabase_url, supabase_api_key)
```

```python
import base64
```

```python
result = supabase.table('hundred_word_articles')\
    .select('id','headline','output','image_url')\
.is_('image_url','null')\
    .order('created_at', desc=False)\
    .execute()

result.data
```

```python
public_urls_3 = []
```

```python
def image_result(image_input):
    for i in image_input:
        image_result = client.images.generate(
            model = "gpt-image-1-mini",
            prompt = f"""
            You are an art director for a modern digital news platform targeting GenZ readers.

            Given a news story, generate an image prompt that:
            - Captures the THEME or EMOTION of the story, not a literal depiction
            - Uses symbolic/conceptual imagery (e.g. for a story about inflation, don't show a chart — show a shrinking shopping cart in a vast empty supermarket)
            - Follows this structure: [symbolic subject], [action/state], [setting], editorial illustration style, [mood-appropriate lighting], [2-3 color palette words]
            - Keeps it under 50 words
            - Never includes text, brand logos, or recognizable faces
            - The image should be in a s
            
            Story headline: {i['headline']}
            Story summary: {i['output']}
            """,
            size = "1024x1024"
        )

        image_base64 = image_result.data[0].b64_json
        image_bytes = base64.b64decode(image_base64)

        file_name = f"{i['id']}.png"

        supabase.storage.from_("article-image").upload(
            file_name,
            image_bytes,
            file_options = {"content-type": "image/png"}
        )

        public_url = supabase.storage.from_("article-image").get_public_url(file_name)
        

        public_urls_3.append({
            'public_url': public_url,
            'id': i['id']
        })

        print(f"✅ {i['headline']} → {public_url}")
```
