```python
supabase: Client = create_client(supabase_url, supabase_api_key)
```

```python
import base64
```

```python
result = supabase.table('hundred_word_articles')\
    .select('id','headline','output')\
    .order('created_at', desc=False)\
    .execute()

result.data
```

```python
def image_result(image_input):
    for i in image_input:
        image_result = client.images.generate(
            model = "gpt-image-1-mini",
            prompt = f"""
            You are a senior visual designer creating a minimal, realistic thumbnail image for a GenZ AI/tech news story.

            Goal:
            Create ONE strong, instantly understandable visual that summarizes the story at a glance.
            
            Hard rules:
            - No headlines
            - No logo of Krux
            - Minimalistic composition, realistic style
            - Avoid clutter: maximum 1 main subject + 1 supporting element
            - Clean background with lots of negative space
            - No collage, no split-screen, no multiple scenes
            - Image SHOULD be in SQUARE
            
            Design requirements:
            - Show a single “hero” object/scene that represents the story’s core idea
            - Use modern, editorial photo style (soft studio lighting, sharp focus) 
            - The image should feel like a premium news app thumbnail
            
            Process:
            1) Identify the story’s core subject (person/company/tech/event) and the main action/conflict.
            2) Convert it into a visual metaphor using real-world objects.

            Here are the details of the news article, both headline and output:

            {i['headline']}

            {i['output']}
            """
        )

        image_base64 = image_result.data[0].b64_json
        image_bytes = base64.b64decode(image_base64)

        file_name = f"{i['id']}"

        with open(file_name, "wb") as f:
            f.write(image_bytes)
```

```python
image_input = [{'id': 54,
  'headline': 'Law Firms Get AI Agent Named Teddy to Handle Lawsuits',
  'output': 'Anytime AI just dropped version 2.0 with "Talk to Teddy"—a chat interface that digs through case files and web research in one encrypted hub. The platform now runs autonomous AI agents that handle multi-step litigation tasks like medical chronologies, demand letters, and discovery responses. It\'s HIPAA-compliant with 256-bit encryption, and your data stays private (no training public models). Targeting personal injury, nursing home negligence, and med-mal cases, CEO Teddy Wu says the goal is ditching clicks for results.\n\nIf autonomous agents nail high-stakes legal work, expect law firms to rapidly adopt AI co-pilots—potentially reshaping junior associate roles and litigation timelines industry-wide.'}]
```

```python
image_result(image_input)
```
