```python
def claude_image_builder(image_prompt_builder):
    final_news_article = []
    for i in image_prompt_builder:
        response = client_anthropic.messages.create(
            model = "claude-sonnet-4-5",
            max_tokens = 85,
            messages = [{
                "role": "user",
                "content":  f"""
                You are an art director for a modern digital news platform.
                Given a news story, generate an image prompt that:
                - Depicts ONE single visual concept that represents the story
                - Maximum 2-3 objects in the entire image
                - Clean, minimal composition with plenty of negative space
                - No charts, no numbers, no percentages, no data visualizations
                - No split compositions or multiple scenes
                - Think "magazine cover" not "infographic"
                - Uses editorial illustration style with bold colors
                - Wide 16:9 format for mobile card thumbnail
                
                Story headline: {i['headline']}
                Story summary: {i['output']}
            """
            }]
        )

        final_output = response.content[0].text

        final_image_dictionary = {
            'final_output': final_output,
            'id': i['id']
        }

        final_news_article.append(final_image_dictionary)

    return final_news_article
```

```python
def image_result(image_input):
    for i in image_input:
        image_result = client.images.generate(
            model = "gpt-image-1-mini",
            prompt = i['final_output'],
            size = "1536x1024"
        )

        image_base64 = image_result.data[0].b64_json
        image_bytes = base64.b64decode(image_base64)

        file_name = f"{i['id']}.png"

        supabase.storage.from_("article-image").upload(
            file_name,
            image_bytes,
            file_options = {"content-type": "image/png"}
        )

        public_url = supabase.storage.from_("article-image").get_public_url(file_name)
        

        public_urls_5.append({
            'public_url': public_url,
            'id': i['id']
        })

        print(f"✅ {i['final_output']} → {public_url}")
```
